```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('../data/sample.csv')
df = df.head(10)
df
```

```python
map = folium.Map(
    location=[
        df['latitude'].mean(),
        df['longitude'].mean()
    ],
    zoom_start=13
)


def get_icon(level, regulation):
    color = ''
    value = level / regulation

    if value < .70:
        color = 'lightblue'
    elif value < 0.99:
        color = 'orange'
    else:
        color = 'red'

    return folium.Icon(
        icon='cloud',
        color=color,
        icon_color='white'
    )

df.apply(
    lambda row: folium.Marker(
        location=[
            row['latitude'],
            row['longitude']
        ],
        popup=row['level'],
        tooltip='<h5>Click here</h5>',
        icon=get_icon(0.8, 1),
    ).add_to(map),
    axis=1)

map
```
